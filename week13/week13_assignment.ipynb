### 2. Data Diversity

Use Python to identify how many unique values each column in your data has.  Use Python to identify which column has the greatest number of distinct values (aka the most diversity)?  If your data has a unique identifier (e.g. a person or visit or record ID) then find the column with the next most number of unique values.

```python
import pandas as pd

# Load the data
data = pd.read_csv('diabetes.csv')

# Find the number of unique values for each column
unique_values = data.nunique()
print("Unique values per column:")
print(unique_values)

# Find the column with the most distinct values
most_unique_column = unique_values.idxmax()
most_unique_values = unique_values.max()

print(f"\nColumn with the most unique values: {most_unique_column} ({most_unique_values} unique values)")
```

### 3. Data Redundancy

Use Python to identify any columns in your data where the value for every row is the same.  For example, if you had a gender column and all the rows said 'M', then you would want to write the code that helped you determine that.

```python
import pandas as pd

# Load the dataframe.
df = pd.read_csv('/content/diabetes.csv')

# Identify columns with constant values
constant_columns = []
for col in df.columns:
    if df[col].nunique() == 1:
        constant_columns.append(col)

# Print the results
if constant_columns:
    print("Columns with constant values:")
    for col in constant_columns:
        print(f"- {col}: {df[col].iloc[0]}")  # Print the constant value
else:
    print("No columns with constant values found.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataframe.
df = pd.read_csv('/content/diabetes.csv')

# Explore the relationship between 'Glucose' and 'Outcome'
plt.figure(figsize=(8, 6))
sns.boxplot(x='Outcome', y='Glucose', data=df)
plt.title('Glucose Levels by Outcome')
plt.xlabel('Diabetes Outcome (0 = No, 1 = Yes)')
plt.ylabel('Glucose Level')
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Diabetes Features')
plt.show()
```

```python
#Further exploration:  Distribution of Age
plt.figure(figsize=(10,6))
sns.histplot(df['Age'], kde=True, bins=20)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Further exploration: Pairplot of key features
sns.pairplot(df[['Glucose', 'BMI', 'Age', 'Outcome']], hue='Outcome')
plt.show()
```
